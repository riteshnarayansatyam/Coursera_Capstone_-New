Importing Numpy and Panda Library

```python
import numpy as np 

import pandas as pd # library for data analsysis

print('Libraries imported.')
```

Importing Data from URL " https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M "

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)

required_cols = ['Postcode', 'Borough','Neighbourhood']

for table in tables:
    if(str(np.array_equal(np.array(table.columns),np.array(required_cols)))=="True"):
        pstl_data_df = pd.DataFrame(table)    
    break
print("Shape of Dataframe is - ",table.shape)
table.head()
```

Removing Data Set where Borough is Not assigned 

```python
#Filtering out Boroughs which are Not assigned
table = table[table.Borough!="Not assigned"]
print("Shape of Dataframe is - ",table.shape)
table.head()
```

```python
table1 = table
table1.head(10)
```

```python
table1 = table1[table1.Neighbourhood!="Not assigned"]
print("Shape of Dataframe is - ",table1.shape)
table1.tail(10)
```

Downloading Data for Latitute and Longitude from URL 'http://cocl.us/Geospatial_data'

```python
one_url = 'http://cocl.us/Geospatial_data'
table2 = pd.read_csv(one_url)
table2.head()
```

Merger of two Tables

```python
table_long = pd.merge(table1, table2, on='Postal Code')
table_long.head()
```

Importing Folium library  

```python
import folium
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(table_long['Latitude'],table_long['Longitude'],table_long['Borough'],table_long['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

Kmean logic is applied . sklearn cluster library is downloaded

```python
from sklearn.cluster import KMeans
k=5
toronto_clustering = table_long.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
table_long.insert(0, 'Cluster Labels', kmeans.labels_)
```

Final Map is prepared

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, neighbourhood, cluster in zip(table_long['Latitude'], table_long['Longitude'], table_long['Neighbourhood'], table_long['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
